```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#This is an upload of the full file
df = pd.read_csv('C:\\Users\\david\\Downloads\\2017-fordgobike-tripdata.csv') 
df
```

```python
#Counts of the start sation Id to get an idea of what Histogram would look like
df.start_station_id.value_counts()
```

```python
#This histogram shows the stations where bike trips originate the most.#This h 
#The top station is station id 15, San Francisco Ferry Building. The Ferry
#Building Station may be a popular spot for tourists who want to bike
#along the Bay
#It shows that a 
#df.hist(by=df['start_station_id'])
plt.hist(df['start_station_id'], bins=10, color='red')
#df['start_station_id']
#plt.hist(df, column='start_station_id', bins=340)
plt.title('Frequency as Origin Station')
plt.xlabel('Station id')

plt.show()
```

```python
#Each data point represents where a trip started and ended.
#It helps to show which stations and trips are the most popular.

plt.figure(figsize=(20, 10))

plt.scatter(
    x=df['start_station_id'],
    y=df['end_station_id'],
    marker= 'x', s=2, color= 'black'
)
plt.show()
```

```python
#This plot shows the length of a trip by user birth year
#The scatter plot shows that younger riders go for longer trips
#There are some large outliers of trips probably because of users
#forgetting to return their bikes to the station

plt.figure(figsize=(20, 10))

plt.scatter(x=df['member_birth_year'],
            y=df['duration_sec'],
           marker= 'x', s=2, color= 'black'
)
plt.title('Duration of Bike Rental by Age')

plt.show()
```
